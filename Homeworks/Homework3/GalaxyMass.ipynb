```python
import numpy as np  #importing numpy library 
import astropy.units as u #importing units from astropy library
from ReadFile import Read #importing our Readfile function 
import pandas as pd #importing pandas as pd
```

```python
#function to calculate total mass of given data with specific filename and particle type 
#Inputs: filename, PType(Particle Type)
#return(s): Mass_sum (total mass)
def ComponentMass(filename,PType):
    file=Read(filename)  #using Read to get all the values from Readfile.py
    time, total, data = Read(filename) #using time, total for number of particles and data to get specifics 
    #about them from our text file
    index = np.where(data['type'] == PType)  #creating an array to store indexes of particles of desired Ptype
    

#NOTE: To convert mass from 1e10 Msun to 10e12 Msun(which is just 1e13), we multiply the mass column by 10e-3

    mnew = data['m'][index]*10e-3*u.Msun  #singling out the column for mass from our data
    df = pd.DataFrame(mnew) #creating a 2-dimensional labeled data structure with column of Mass m 
    Mass_sum=df.sum() #summing up our dataframe to show total mass for given particule type
    Mass_sum = np.round(Mass_sum,3)# Value is rounded to 3 decimal places
    
    
        
    return Mass_sum

```

```python
#NOTE: There's a 0 in the assigned row because pd.Datafram creates a 2 dimensional structure
# the relevant values are on the right
#Calling our function for given 3 text files to get Halo, Disk and Bulge mass for specific galaxies
Halo_Mass_Milky=ComponentMass("MW_000.txt",1) 
Disk_Mass_Milky=ComponentMass("MW_000.txt",2)
Bulge_Mass_Milky=ComponentMass("MW_000.txt",3)
Total_Mass_Milky=Halo_Mass_Milky+Disk_Mass_Milky+Bulge_Mass_Milky
print("Halo Mass of milky way:",Halo_Mass_Milky)
print("Disk Mass of milky way:",Disk_Mass_Milky)
print("Bulge Mass of milky way:",Bulge_Mass_Milky)
print("Total Mass of milky way:" ,Total_Mass_Milky)
Halo_Mass_M31=ComponentMass("M31_000.txt",1)
Disk_Mass_M31=ComponentMass("M31_000.txt",2)
Bulge_Mass_M31=ComponentMass("M31_000.txt",3)
Total_Mass_M31=Halo_Mass_M31+Disk_Mass_M31+Bulge_Mass_M31
print("Halo Mass of M31:",Halo_Mass_M31)
print("Disk Mass of M31:", Disk_Mass_M31)
print(" Bulge Mass of M31:",Bulge_Mass_M31)
print("Total Mass of M31:" ,Total_Mass_Milky)
Halo_Mass_M33=ComponentMass("M33_000.txt",1)
Disk_Mass_M33=ComponentMass("M33_000.txt",2)
Bulge_Mass_M33=ComponentMass("M33_000.txt",3)
Total_Mass_M33=Halo_Mass_M33+Disk_Mass_M33+Bulge_Mass_M33
print("Halo Mass of M33:",Halo_Mass_M33)
print("Disk Mass of M33:",Disk_Mass_M33)
print("Bulge Mass of M33:",Bulge_Mass_M33)
print("Total Mass of M33:",Total_Mass_M33)
#Calculating the Total Halo, Disk and Bulge mass for all 3 galaxies combined, called the local group
Local_Group_Halo=Halo_Mass_Milky+Halo_Mass_M31+Halo_Mass_M33
print("Local Group Halo Mass",Local_Group_Halo)
Local_Group_Disk=Disk_Mass_Milky+Disk_Mass_M31+Disk_Mass_M33
print("Local Group Disk Mass",Local_Group_Disk)
Local_Group_Bulge=Bulge_Mass_Milky+Bulge_Mass_M31+Bulge_Mass_M33
print("Local Group Bulge Mass",Local_Group_Bulge)
Local_Group_Total=Local_Group_Halo+Local_Group_Disk+Local_Group_Bulge
print("Local Group Total Mass",Local_Group_Total)
#Dividing the stellar mass(in the form of Disk and Bulge mass) by the total Mass to get the Baryon function
#For specific galaxies and the local group as a whole , rounded to 3 decimal places
f_bar_milky=np.round((Disk_Mass_Milky+Bulge_Mass_Milky)/Total_Mass_Milky,3)
f_bar_M31=np.round((Disk_Mass_M31+Bulge_Mass_M31)/Total_Mass_M31,3)
f_bar_M33=np.round((Disk_Mass_M33+Bulge_Mass_M33)/Total_Mass_M33,3)
f_bar_LG=np.round((Local_Group_Disk+Local_Group_Bulge)/Local_Group_Total,3)
print("Milky way baryon fraction",f_bar_milky)
print("M31 baryon fraction",f_bar_M31)
print("M33 baryon fraction",f_bar_M33)
print("Local Group baryon fraction",f_bar_LG)
```
